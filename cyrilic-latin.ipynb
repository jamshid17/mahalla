```python
translation_list = {} 
```

```python
for key, value in translation_list_upper.items():
    translation_list[key] = value
    key_lower = key.lower()
    value_lower = value.lower()
    translation_list[key_lower] = value_lower
    
```

```python
def translate(word):
    new_word = ''
    word = word.replace("\n", " ").replace("  ", " ").replace('"', '')
    while True:
        if word[0] == ' ':
            word = word[1:]
        elif word[-1]  == ' ':
            word = word[:-1]
        else:
            break
        
    for letter in word:
        if letter in translation_list.keys():
            latin_letter = translation_list[letter]
            new_word += latin_letter
        else:
            print(letter, word)
            raise Exception("Sorry")
    return new_word
```
